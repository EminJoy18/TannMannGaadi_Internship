## Matplotlib is the most commonly used library for plotting purposes

Typically, two APIs are called :
1. Global API
2. Object Oriented API

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## GLOBAL API

Matplotlib's default pyplot API has a global, MATLAB-style interface

```python
x = np.arange(-10,11)

plt.figure( figsize = (12,6) )   # size of the plot
plt.title("My Nice Plot")

plt.plot( x , (x**2) )
plt.plot( x , -(x**2) )
```

```python
plt.figure( figsize=(12,6) )
plt.title("Graph of y-x^2")


plt.subplot(1, 3, 1)
# -> rows, column and panel selected
# 1, 2, 1 => my plot will have 1 row, 2 columns and 1st panel


plt.plot(x, x**2)           # -> The parabolic curve
plt.plot( [0,0,0] , [-10,0,100] ) # -> The straight line -> The line joining the points (0,0) and (0,80)  -> It can be in shape of (3,) as well
plt.xlabel("X")
plt.ylabel("X - squared")
plt.legend([ 'X^2' , 'Vertical Line' ])


plt.subplot(1, 3, 3)
# 1, 2, 2 => my plot will have 1 row, 2 columns and 2nd panel

plt.plot( x , -(x**2) )
plt.plot( [-10,0,10] , [ -50, -50 , -50] , 'mo--' )  # -> has formatting options as well
# these coordinates tell u the span of the line
# the coordinaes of joining are : (-10,-50) (0,-50) (10,-50)

# g = green , o = bold circle markers , -- = dashed line

plt.xlabel("X")
plt.ylabel("Negative of X-Squared")
plt.legend( [ '-(X^2)' , 'Vertical Line' ] )
```

```python
plt.plot( [-10,0,10] , [-50,-50,-50] , 'b--')
```

## **OOP Interface**

for a figure with multiple axes

```python
plot_objects = plt.subplots( nrows=2 , ncols = 2 , figsize=(20,10) )
# we directly use suplots function

fig, ((ax1,ax2),(ax3,ax4)) = plot_objects

plot_objects
```

```python
ax1.plot(np.random.rand(50), 'b--')
ax2.plot(np.random.rand(50))
ax4.plot(np.random.rand(50) , 'o')   # scatter plot -> just the dots
ax3.plot(np.random.rand(50) , marker='o' , linewidth = 3.0)

fig
```

### **subplot2grid** command

```python
plt.figure(figsize=(14,6))

plt.subplot2grid( (3,3) , (0,0) , colspan=3 ).plot(np.random.rand(10))
plt.subplot2grid( (3,3) , (1,0) , colspan=2 )
plt.subplot2grid( (3,3) , (1,2) , rowspan=3 ).plot(np.random.rand(5) , 'mo')
plt.subplot2grid( (3,3) , (2,0))
plt.subplot2grid( (3,3) , (2,1)).plot(np.random.rand(5) , 'go--')
```

### **Scatter plots**

```python
N = 50

x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * ( 20 * np.random.rand(N)) ** 2    # 0 to 15 point radii
```

```python
plt.figure(figsize=(12,6))

plt.scatter( x,y, s=area, c=colors , alpha=0.5 ,cmap='Spectral')
# alpha = the transparency of the plotted circles
# s = area => the size of the plotted circle is different so that it signifies something ; otherwise, all circles would be dots
# cmap (colormaps) = 'Spectral' => the color of plotted circles matches according to some wavelength or something
# c => array or list of colors


plt.colorbar()

plt.show()
```

Here, in the above graph we are easily encoding 4 dimensions.

X , Y , color of circle , size of plotted circle

```python
fig = plt.figure(figsize=(14, 6))

ax1 = fig.add_subplot(1,2,1)
plt.scatter(x, y, s=area, c=colors, alpha=1.0, cmap='Pastel1')
plt.colorbar()

ax2 = fig.add_subplot(1,2,2)
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Pastel2')
plt.colorbar()

plt.show()
```

## **HISTOGRAM**

```python
values = np.random.randn(1000)
```

In the 'histtype' parameter of hst function there are 4 options:
1. bar
2. barstacked
3. step        (like silhouette)
4. stepfilled

```python
plt.subplots(figsize=(12,6))

plt.hist(values, bins=100, alpha=0.8,
         histtype='bar', color='steelblue',
         edgecolor='green')
# bins => affect the size of the bar

plt.xlim(xmin=-5,xmax=5)  # offset from either sides

plt.show()
```

```python
# fig.savefig('hist.png')   # to save the histogram as an image
```

## **KERNEL DENSITY ESTIMATION**

```python
from scipy import stats
```

```python
density = stats.kde.gaussian_kde(values)
density
```

```python
plt.figure( figsize = (12,6) )

values2 = np.linspace( min(values)-10 , max(values)+10 , 100 )
plt.plot(values2 , density(values2) , color='#FF7F00')
plt.fill_between(values2, 0, density(values2) , alpha=0.5 , color='#FF7F00')
plt.xlim(xmin=-5 , xmax=5)

plt.show()
```

### **COMBINE PLOTS**

```python
plt.figure(figsize=(12,6))

plt.hist(values, bins=100, alpha=0.8, density=1,
         histtype='bar', color='steelblue',
         edgecolor='green')

plt.plot(values2 , density(values2) , color='#FF7F00', linewidth=2.0)

plt.xlim(xmin=-5,xmax=5)

plt.show()
```

### **BAR PLOTS**

```python
y1 = np.random.rand(1,5)[0]
y2 = np.random.rand(1,5)[0]
```

```python
plt.figure(figsize=(10,5))

barwidth = 0.5
plt.bar(np.arange(len(y1)) , y1 , width = barwidth , color='#00b894')

plt.show()
```

Bars can also be stacked

```python
plt.figure(figsize=(10,5))

barwidth = 0.5
plt.bar(np.arange(len(y2)) , y2 , width = barwidth , color='#e17055', label='label y2')
plt.bar(np.arange(len(y1)) , y1 , width = barwidth , color='#00b894', label='label y1')

plt.legend()
plt.show()
```
